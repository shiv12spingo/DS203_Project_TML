# **Worldwide**

```
import pandas as pd
gtd=pd.read_excel('/content/drive/MyDrive/globalterrorismdb_0919dist.xlsx')
display(gtd)
```

### **Top 10 cities prone to terrorism**

```
import seaborn as sn
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(15,8))
gtd2=gtd[gtd['city']!='Unknown']
cities=gtd2['city'].value_counts()[:10]
ax=sn.barplot(x=cities.index,y=cities.values)
ax.set_title('Most affected cities')
ax.set_xlabel('Cities')
ax.set_ylabel('Number of attacks')
```

### **Top 10 countries prone to terrorism**

```
fig,ax=plt.subplots(figsize=(15,8))
countries=gtd['country_txt'].value_counts()[:10]
ax=sn.barplot(x=countries.index,y=countries.values)
plt.title('Most affected countries')
ax.set_xlabel('Countries')
ax.set_ylabel('Number of attacks')
```

### **Top 10 regions prone to terrorism**

```
fig,ax=plt.subplots(figsize=(24,12))
regions=gtd['region_txt'].value_counts()[:10]
ax=sn.barplot(x=regions.index,y=regions.values)
plt.title('Most affected regions')
ax.set_xlabel('Regions')
ax.set_ylabel('Number of attacks')
```

### **Top 5 popular weapons**

```
fig,ax=plt.subplots(figsize=(12,6))
gtd4=gtd[gtd['weaptype1_txt']!='Unknown']
weapons=gtd4['weaptype1_txt'].value_counts()[:5]
ax=sn.barplot(x=weapons.index,y=weapons.values)
plt.title('Most popular weapons')
ax.set_xlabel('Weapons')
ax.set_ylabel('Number of attacks')
```

### **5 Most frequent modes of attack**

```
fig,ax=plt.subplots(figsize=(14,8))
gtd4=gtd[gtd['attacktype1_txt']!='Unknown']
modes=gtd4['attacktype1_txt'].value_counts()[:5]
ax=sn.barplot(x=modes.index,y=modes.values)
plt.title('Most popular modes')
ax.set_xlabel('Type of attack')
ax.set_ylabel('Number of attacks')
```

### **5 Most imminent targets**

```
fig,ax=plt.subplots(figsize=(14,8))
gtd4=gtd[gtd['targtype1_txt']!='Unknown']
targets=gtd4['targtype1_txt'].value_counts()[:5]
ax=sn.barplot(x=targets.index,y=targets.values)
plt.title('Most frequent targets')
ax.set_xlabel('Categories of people')
ax.set_ylabel('Number of attacks')
```

### **5 Most active terrorist organisations**



```
fig,ax=plt.subplots(figsize=(16,10))
gtd4=gtd[gtd['gname']!='Unknown']
orgs=gtd4['gname'].value_counts()[:5]
ax=sn.barplot(x=orgs.index,y=orgs.values)
plt.title('Most active terror groups')
ax.set_xlabel('Terrorist Organisations')
ax.set_ylabel('Number of attacks')
```

### **5 Most deadly attacks**

```
cols=['iyear','country_txt', 'region_txt', 'provstate', 'city', 'suicide', 'attacktype1_txt', 'targtype1_txt', 'gname', 'weaptype1_txt', 'nkill', 'nwound']
gtd3=gtd[cols]
dead=gtd3.sort_values(by='nkill', ascending=False)
dead.head(5)
```
